2 baris di bawah biar stasioner karena time series harus stasioner

```python
from statsmodels.tsa.stattools import adfuller
print('Results of Dickey-Fuller Test:')
dftest = adfuller(data_chile['PercentOfBaseline'], autolag='AIC')
dfoutput = pd.Series(dftest[0:4], index=['Test Statistic','p-value','#Lags Used','Number of Observations Used'])
for key,value in dftest[4].items():
    dfoutput['Critical Value (%s)'%key] = value
print(dfoutput)
```

```python
#define function for kpss test
from statsmodels.tsa.stattools import kpss
#define KPSS
print ('Results of KPSS Test:')
kpsstest = kpss(data_chile['PercentOfBaseline'], regression='c')
kpss_output = pd.Series(kpsstest[0:3], index=['Test Statistic','p-value','Lags Used'])
for key,value in kpsstest[3].items():
    kpss_output['Critical Value (%s)'%key] = value
print (kpss_output)

```

```python
#define function for kpss test
from statsmodels.tsa.stattools import kpss
#define KPSS
print ('Results of KPSS Test:')
kpsstest = kpss(data_chile['diff'].dropna(), regression='c')
kpss_output = pd.Series(kpsstest[0:3], index=['Test Statistic','p-value','Lags Used'])
for key,value in kpsstest[3].items():
    kpss_output['Critical Value (%s)'%key] = value
print (kpss_output)
```

```python
training_set=data_chile
training_set=pd.DataFrame(training_set)
```

```python
training_set
```

```python
training_set=training_set.drop(columns=['diff'])
```

```python
training_set
```

```python
# Feature Scaling
from sklearn.preprocessing import MinMaxScaler
sc = MinMaxScaler(feature_range = (0, 1))
training_set_scaled = sc.fit_transform(training_set)
```

```python
training_set_scaled
```

```python
# training_set_scaled=training_set_scaled.nonzero()
```

```python
training_set_scaled
```

```python
# Creating a data structure with 60 timesteps and 1 output
X_train = []
y_train = []
for i in range(60, 210):
    X_train.append(training_set_scaled[i-60:i, 0])
    y_train.append(training_set_scaled[i, 0])
```

```python
X_train, y_train = np.array(X_train), np.array(y_train)
```

```python
# Reshaping
X_train = np.reshape(X_train, (X_train.shape[0], X_train.shape[1], 1))
```

```python
X_train   
```

```python
y_train
```

```python
#Building the RNN
#Importing the Keras libraries and packages
from keras.models import Sequential
from keras.layers import Dense
from keras.layers import LSTM
from keras.layers import Dropout
```

```python
# Initialising the RNN
model = Sequential()
```

```python


model.add(LSTM(units = 80,return_sequences = True,input_shape = (X_train.shape[1],1)))
model.add(Dropout(0.2))
#unit 50 jadi 80
#dropout terakhir 0.1
model.add(LSTM(units = 80,return_sequences = True))
model.add(Dropout(0.2))

model.add(LSTM(units = 80,return_sequences = True))
model.add(Dropout(0.2))



model.add(LSTM(units = 80))
model.add(Dropout(0.2))

model.add(Dense(units = 1))
```

```python
# Compiling the RNN
# Code here
#adam jadi SGD
model.compile(optimizer = 'adam',loss = 'mean_squared_error')
```

```python
# Fitting the RNN to the Training set
epoch = 100
batch_size = 42
#model.fit(#code here)
model.fit(X_train,y_train,epochs = epoch, batch_size = batch_size, verbose=1)
```
