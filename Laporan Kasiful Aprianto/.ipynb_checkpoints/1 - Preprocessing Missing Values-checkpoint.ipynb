```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv("./covid_impact_on_airport_traffic.csv")
```

```python
df.head()
```

```python
df = df[
    ['Date', 'AirportName',
       'PercentOfBaseline', 'City', 'State', 'Country']
]
```

```python
df.head()
```

```python
df.AirportName.unique()
```

```python
dset = pd.DataFrame({
    "Date": pd.date_range(start=df.Date.min(), end=df.Date.max(), freq="d")
})
```

```python
for airport in df.AirportName.unique():
    temp = df.loc[df.AirportName == airport, ['Date', 'PercentOfBaseline']]
    temp['Date'] = pd.to_datetime(temp.Date)
    dset = dset.merge(temp, how='left', on='Date')
    dset.columns = dset.columns.str.replace('PercentOfBaseline', airport)
```

```python
dset
```

```python
for kol in dset.columns[1:]:
    print(kol, "==> ", any(pd.isna(dset[kol])))
```

```python
for airport in dset.columns[1:]:
    temp = dset[airport]
    temp = temp.interpolate()
    dset[airport] = temp
```

```python
dset.to_excel("traffic_v2.xlsx")
```

```python
negara = df.Country.unique()
```

```python
dset_per_negara = pd.DataFrame({"Date": pd.date_range(start=df.Date.min(), end=df.Date.max(), freq="d")})
```

```python
for neg in negara:
    list_bandara = df.loc[df.Country ==neg, 'AirportName'].unique()
    temp = dset[list_bandara]
    temp1 = pd.DataFrame({"Date": pd.date_range(start=df.Date.min(), end=df.Date.max(), freq="d")})
    temp1[neg] = temp.sum(axis=1)
    dset_per_negara = dset_per_negara.merge(temp1, how='left', on='Date')
    # dset.columns = dset.columns.str.replace('PercentOfBaseline', airport)
```
